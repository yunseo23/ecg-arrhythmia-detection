```python
import wfdb
import neurokit2 as nk
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks

# 설정
record = "219"  # MIT-BIH record 중 하나
fs = 360        # 샘플링 주파수
length = 10000   # 시각화 구간 길이

# 1. 원본 ECG 로드
signal, _ = wfdb.rdsamp(f'mitdb/{record}', channels=[0])
signal = np.squeeze(signal)

# 2. ecg_clean() 적용
cleaned = nk.ecg_clean(signal, sampling_rate=fs, method='biosppy')

# 3. ecg_peaks() 적용
_, rpeaks_info = nk.ecg_peaks(cleaned, sampling_rate=fs, method='neurokit')
rpeaks = rpeaks_info['ECG_R_Peaks']

# 4. adjust_rpeaks 적용 함수 정의
def adjust_rpeaks(sig, rpeaks, height_threshold=0.4):
    adjusted = rpeaks.copy()
    for i in range(len(rpeaks) - 1):
        segment = sig[rpeaks[i]:rpeaks[i+1]]
        peak_height = (np.max(sig) - np.min(sig)) * height_threshold
        peaks, _ = find_peaks(segment, height=peak_height)
        if len(peaks) > 0:
            adjusted[i] = rpeaks[i] + peaks[0]  # 첫 번째 peak 사용
    return adjusted

adjusted_rpeaks = adjust_rpeaks(cleaned, rpeaks)

# 시각화 1: 원본 시그널
plt.figure(figsize=(15, 3))
plt.plot(signal[:length], label="Original ECG")
plt.title("1. Original ECG Signal")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()

# 시각화 2: 정제된 시그널 (ecg_clean)
plt.figure(figsize=(15, 3))
plt.plot(cleaned[:length], label="Cleaned ECG", color='orange')
plt.title("2. Cleaned ECG Signal")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()

# 시각화 3: Cleaned + R-peaks (ecg_peaks)
plt.figure(figsize=(15, 3))
plt.plot(cleaned[:length], label="Cleaned ECG", color='green')
rpeaks_plot = [rp for rp in rpeaks if rp < length]
plt.plot(rpeaks_plot, cleaned[rpeaks_plot], 'bo', label='Detected R-peaks')
plt.title("3. Cleaned ECG with R-peaks (ecg_peaks)")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()

# 시각화 4: Cleaned + R-peaks + Adjusted (adjust_rpeaks)
plt.figure(figsize=(15, 3))
plt.plot(cleaned[:length], label="Cleaned ECG", color='purple')
adj_plot = [rp for rp in adjusted_rpeaks if rp < length]
plt.plot(adj_plot, cleaned[adj_plot], 'ro', label='Adjusted R-peaks')
plt.title("4. Cleaned ECG with Adjusted R-peaks")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()

# --- 시각화 5: 정제된 ecg_peaks + adjust_rpeaks ---
plt.figure(figsize=(15, 3))
plt.plot(cleaned[:length], label="Cleaned ECG", color='green')
rpeaks_plot = [rp for rp in adjusted_rpeaks if rp < length]
plt.plot(rpeaks_plot, cleaned[rpeaks_plot], 'ro', label="Adjusted R-peaks")
plt.title("5. ecg_peaks + Adjusted R-peaks")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from src.preprocess.data_loader import get_mitdb_records, load_ECG_signal, load_symbols
from src.preprocess.signal_process import ecg_clean, get_rpeaks, adjust_rpeaks
from src.preprocess.label_process import extract_labels, group_labels, group_labels_binary
from config import MITDB_PATH, HYPERPARAMS
import pandas as pd
import numpy as np
from tqdm import tqdm

# 설정
FS = HYPERPARAMS['fs']
EX_LABELS = HYPERPARAMS['ex_labels']

original_labels = []
aami_labels = []
binary_labels = []

mitdb = get_mitdb_records()

for record in tqdm(mitdb):
    # 1. Load ECG + annotations
    sig = load_ECG_signal(record)
    dct_symbols = load_symbols(record, MITDB_PATH, extension='atr', EX_LABELS=EX_LABELS)

    # 2. Clean and detect R-peaks
    sig_cleaned = ecg_clean(sig, fs=FS, method='biosppy')
    rpeaks = get_rpeaks(sig_cleaned, ecg_peaks_method='neurokit')
    adj_rpeaks, _ = adjust_rpeaks(sig_cleaned, rpeaks)

    # 3. Label extraction
    labels = extract_labels(adj_rpeaks, dct_symbols)
    original_labels.extend(labels)
    aami_labels.extend(map(group_labels, labels))
    binary_labels.extend(map(group_labels_binary, labels))

# --- AAMI 기준 분포 테이블 ---
aami_df = pd.DataFrame({
    "Original Label": original_labels,
    "AAMI Group": aami_labels
})
aami_summary = (
    aami_df
    .groupby(["Original Label", "AAMI Group"])
    .size()
    .reset_index(name='Count')
)
total = aami_summary['Count'].sum()
aami_summary['Percentage'] = (aami_summary['Count'] / total * 100).round(2)

# --- Binary 분류 테이블 ---
binary_series = pd.Series(binary_labels, name='Binary Label')
binary_summary = (
    binary_series
    .value_counts()
    .rename_axis('Label')
    .reset_index(name='Count')
)
binary_total = binary_summary['Count'].sum()
binary_summary['Percentage'] = (binary_summary['Count'] / binary_total * 100).round(2)

# --- 콘솔에 출력 ---
print("\n📊 AAMI Group Distribution:")
print(aami_summary)

print("\n📊 Binary Classification (S vs Non-S) Distribution:")
print(binary_summary)
```
